```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas_ta as ta

# Load data
df = pd.read_csv(
    'data/^GSPC.csv',
    index_col=0,
    parse_dates=True,
    skiprows=[1]  # keep if your CSV has an extra header row
)
df.index.name = 'Date'
df = df.apply(pd.to_numeric, errors='coerce')

# Calculate SMA & RSI
df['SMA50'] = df['Close'].rolling(50).mean()
df['RSI'] = ta.rsi(df['Close'], length=14)

# Plot closing price + SMA50
df[['Close', 'SMA50']].plot(title='S&P 500 Closing Price + SMA50')
plt.xlabel('Date')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot RSI with overbought/oversold lines
df['RSI'].plot(title='RSI')
plt.axhline(70, color='red', linestyle='--', label='Overbought (70)')
plt.axhline(30, color='green', linestyle='--', label='Oversold (30)')
plt.xlabel('Date')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


```
